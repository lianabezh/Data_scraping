```python
import re 
import time
import requests
import numpy as np
import pandas as pd
from textblob import TextBlob , Word
from scrapy.http import TextResponse
```

```python
url = "https://www.imdb.com/chart/top"
page = requests.get(url)
response = TextResponse(url= page.url, body = page.text , encoding = "utf-8")
```

```python
name =response.css("td[class='titleColumn']> a::text").extract()
```

```python
response.css("span[class='secondaryInfo']::text").extract()
year = response.css("span[class='secondaryInfo']::text").re("\d{4}")
```

```python
rating = response.css("td[class='ratingColumn imdbRating']> strong::text").extract()
```

```python
href = response.css("td[class='titleColumn']> a::attr(href)").extract()
```

```python
print(name, "\n", year, "\n", href , "\n" , rating )
```
