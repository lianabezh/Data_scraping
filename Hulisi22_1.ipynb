```python
def scrap_info(url):
    page=requests.get(url)
    response=TextResponse(url= page.url, body = page.text , encoding ="utf-8")
    address =response.css("strong[clss='addr']::text").extract_first()
    size = re.sub('[\n\t]',"", response.xpath("//span[@class='ruler']/text()").extract()[0])
    size=re.findall("\d+", size)[0]
    try:
        price=response.xpath("//div[@class = 'price-w']/text()").re("\d.*\d")[0]
    except:
        price=""
    return address,size, price
```

```python
url="https://www.estate.am/"

page=requests.get(url)
response=TextResponse(url= page.url, body = page.text , encoding ="utf-8")
base = "https://www.estate.am"
```
